## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
df_school_data = pd.read_csv("Resources/schools_complete.csv")
df_school_data
```

```python
df_student_data = pd.read_csv("Resources/students_complete.csv")
df_student_data
```

```python
df_CompleteSchool_data = pd.merge(df_student_data, df_school_data, how="left", on=["school_name", "school_name"])
df_CompleteSchool_data
```

```python
df_CompleteSchool_data.describe()
```

```python
# Calculate the total number of schools
# totalSchools = school_data_complete["school_name"].nunique()
TotalSchools = df_CompleteSchool_data["school_name"].nunique()
TotalSchools
```

```python
# Calculate the total number of students
TotalStudents = df_CompleteSchool_data["Student ID"].nunique()
TotalStudents
```

```python
# Calculate the total budget
TotalBudget = df_CompleteSchool_data["budget"].sum()
TotalBudget
```

```python
# Calculate the average math score
AverageMathScore = df_CompleteSchool_data["math_score"].mean()
AverageMathScore
```

```python
# Calculate the average reading score
AverageReadingScore = df_CompleteSchool_data["reading_score"].mean()
AverageReadingScore
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
PassingMath = len(df_CompleteSchool_data[df_CompleteSchool_data["math_score"] >= 70])
PercentagePassingMath = PassingMath / TotalStudents * 100
PercentagePassingMath
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
PassingReading = len(df_CompleteSchool_data[df_CompleteSchool_data["reading_score"] >= 70]) 
PercentagePassingReading = PassingReading / TotalStudents * 100
PercentagePassingReading
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing)
PercentageOverallPassingRate = (AverageMathScore + AverageReadingScore) / 2
PercentageOverallPassingRate
```

```python
# Create a dataframe to hold the above results

District_Summary_df = pd.DataFrame({
    "Total Schools": [TotalSchools],
    "Total Students": [TotalStudents],
    "Total Budget": [TotalBudget],
    "Average Math Score": [AverageMathScore],
    "Average Reading Score": [AverageReadingScore],
    "% Passing Math": [PercentagePassingMath],
    "% Passing Reading": [PercentagePassingReading],
    "% Overall Passing Rate": [PercentageOverallPassingRate]
})
District_Summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# School Name
GroupedSchool_df = df_CompleteSchool_data.groupby(["school_name"])
GroupedSchool_df
```

```python
# School Type
SchoolType = GroupedSchool_df["type"].first()
SchoolType
```

```python
# Total Students
TotalStudents = GroupedSchool_df["Student ID"].count()
TotalStudents
```

```python
# Total School Budget
TotalSchoolBudget = GroupedSchool_df["budget"].first()
TotalSchoolBudget
```

```python
# Per Student Budget
PerStudentBudget = TotalSchoolBudget / TotalStudents
PerStudentBudget
```

```python
# Average Math Score
AverageMathScore = GroupedSchool_df["math_score"].mean()
AverageMathScore
```

```python
# Average Reading Score
AverageReadingScore = GroupedSchool_df["reading_score"].mean()
AverageReadingScore
```

```python
# % Passing Math
PassingMath = df_CompleteSchool_data[df_CompleteSchool_data["math_score"] >= 70].groupby(["school_name"])["math_score"].count()
PercentagePassingMath = PassingMath / TotalStudents * 100
PercentagePassingMath
```

```python
# % Passing Reading
PassingRead = df_CompleteSchool_data[df_CompleteSchool_data["reading_score"] >= 70].groupby(["school_name"])["reading_score"].count()
PercentagePassingReading = PassingRead / TotalStudents * 100
PercentagePassingReading 
```

```python
# Overall Passing Rate (Average of the above two)
PercentageOverallPassingRate = (PercentagePassingMath + PercentagePassingReading) / 2
PercentageOverallPassingRate
```

```python
# Create a dataframe to hold the above results
SchoolSummary_df = pd.DataFrame({"School Type": SchoolType,
      "Total Students": TotalStudents,
      "Total School Budget": TotalSchoolBudget,
      "Per Student Budget": PerStudentBudget,
      "Average Math Score": AverageMathScore,
      "Average Reading Score": AverageReadingScore,
      "% Passing Math": PercentagePassingMath,
      "% Passing Reading": PercentagePassingReading,
      "% Overall Passing Rate": PercentageOverallPassingRate})
SchoolSummary_df
```

## Top Performing Schools (By % Overall Passing)

```python
# Sort and display the top five performing schools by % overall passing.
SchoolSummary_df = SchoolSummary_df.sort_values(["% Overall Passing Rate"], ascending=False)
SchoolSummary_df.head()
```
